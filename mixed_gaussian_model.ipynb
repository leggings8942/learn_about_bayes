```python
%load_ext autoreload
%autoreload 2

from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from Mixed_Gaussian_Model import My_K_means, My_Mixed_Gaussian_Model_in_VB, My_Mixed_Gaussian_Model_in_GS

from Mixed_Gaussian_Model import My_Mixed_Gaussian_Model_with_Variance_in_VB
from Mixed_Gaussian_Model import My_Mixed_Gaussian_Model_with_Variance_in_GS
```

```python
# サンプルデータの作成
np.random.seed(0)
sample_num  = 500
sample_data = np.concatenate([np.random.normal(loc=(4*i, 4*j), scale=1, size=(sample_num, 2)) for i in range(1, 3) for j in range(1, 3)])

num_clusters = int(sample_data.shape[0] / sample_num)
sample_data.shape
```

```python
# KMeansモデルの学習
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(sample_data)

# 各パラメータの取得
labels  = kmeans.predict(sample_data)
centers = kmeans.cluster_centers_

# KMeansの結果を可視化
plt.scatter(sample_data[:, 0], sample_data[:, 1], c=labels, s=5)
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75)
plt.title('KMeans Clustering')
plt.show()
```

```python
# 混合ガウスモデルの学習
model = GaussianMixture(n_components=num_clusters)
model.fit(sample_data)

# 各パラメータの取得
labels  = model.predict(sample_data)
centers = model.means_


# 結果の可視化
plt.scatter(sample_data[:, 0], sample_data[:, 1], c=labels, s=5)
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75)
plt.title('Gaussian Mixture Model')
plt.show()
```

```python
# 自作のKMeansクラスを使用してクラスタリングを行う
model = My_K_means(n_clusters=num_clusters)
model.fit(sample_data)

# 各パラメータの取得
labels  = model.predict(sample_data)
centers = model.centers


# 結果の可視化
plt.scatter(sample_data[:, 0], sample_data[:, 1], c=labels, s=5)
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75)
plt.title('Gaussian Mixture Model')
plt.show()
```

```python
# 自作の混合ガウスモデルを使用してクラスタリングを行う
model = My_Mixed_Gaussian_Model_in_VB(n_clusters=num_clusters)
model.fit(sample_data)

# 各パラメータの取得
labels  = model.predict(sample_data)
centers = model.cluster_μ_k

# 結果の可視化
plt.scatter(sample_data[:, 0], sample_data[:, 1], c=labels, s=5)
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75)
plt.title('Gaussian Mixture Model in VB')
plt.show()
```

```python
# 自作の混合ガウスモデルを使用してクラスタリングを行う
model = My_Mixed_Gaussian_Model_in_GS(n_clusters=num_clusters)
model.fit(sample_data)

# 各パラメータの取得
labels  = model.predict(sample_data)
centers = model.cluster_μ_k

# 結果の可視化
plt.scatter(sample_data[:, 0], sample_data[:, 1], c=labels, s=5)
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75)
plt.title('Gaussian Mixture Model in VB')
plt.show()
```
